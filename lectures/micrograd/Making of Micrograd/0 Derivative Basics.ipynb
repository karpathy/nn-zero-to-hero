# Basics of Derivative

```python
# Important libraries to import to perform further operations

import math                     # for performing mathematical operations
import numpy as np              # for numerical computing that provides support for arrays, mathematical functions, etc.
import matplotlib.pyplot as plt # for creating static, interactive, and animated visualizations.

# display matplotlib plots directly within the notebook interface.
%matplotlib inline              
```

```python
# example function: 3x^2 âˆ’ 4x + 5
def f(x):
    return 3*x**2 - 4*x + 5
```

```python
# when input is 3 (int)
print(f(3))
```

```python
# when input is 3.0 (float)
print(f(3.0))
```

```python
# other input e.g.
print(f(1))
print(f(0))
print(f(-1))
print(f(100))
```

```python
# array of evenly spaced values within (-5,5)
xs = np.arange(-5,5,0.25)
print(xs)
print("xs type:", type(xs))
print(len(xs))
```

```python
# Applying func to each scalar in xs
ys = f(xs)
print(ys)
print(type(ys))
```

```python
# plotting all the inputs xs using matplotlib 
print(plt.plot(xs))
```

```python
# plotting all the outputs ys
print(plt.plot(ys))
```

```python
# Plot ys versus xs (notice the x-axis values in this case (-5, 5))
plt.plot(xs, ys)

# Turn on the grid
plt.grid(True)

# Display the plot
plt.show()
```

# Derivative:

![Derivative](https://miro.medium.com/v2/resize:fit:828/format:webp/0*6-cw21HL-I0hXxjF.png)

```python
# small increment value 'h' which we will add to 'x' for computing the derivative at x=3
h = 0.00000001
x = 3.0
```

```python
# output of function 3x^2 âˆ’ 4x + 5 when x=3
print(f(x))

# output of function 3x^2 âˆ’ 4x + 5 when x=3.00000001
print(f(x+h))
```

```python
# difference
print(f(x+h) - f(x))
```

```python
# derivative
print((f(x+h) - f(x)) / h)
```

## Performing Manual Derivative

Derivative of \(3x^2 - 4x + 5\) is \(6x - 4\).


```python
# therefore:  6x - 4 = 6*3 - 4 = 14
print(6*3.0 - 4)
```

### derivative of function with multiple inputs

```python
# small increment
h = 0.0001

# inputs
a = 2.0
b = -3.0
c = 10.0

# function
d1 = a*b + c

# increasing a with small amount
a += h 

# function after increasing 'a' with small amount
d2 = a*b + c


print("d1:", d1)
print("d2:", d2)
print("slope:", (d2-d1)/h)
```

## since: derivative of (a*b + c) w.r.t a ----> b

```python
h = 0.0001

# inputs
a = 2.0
b = -3.0
c = 10.0

# function evaluation 
d1 = a*b + c

# increasing b with small amount
b += h 

# function evaluation after increasing b with a small amount
d2 = a*b + c


print("d1:", d1)
print("d2:", d2)
print("slope:", (d2-d1)/h)
```

## since: derivative of (a*b + c) w.r.t b ----> a
## here negative sign indicates that the function is decreasing at that point

```python
h = 0.0001

# inputs
a = 2.0
b = -3.0
c = 10000.0

# function evaluation 
d1 = a*b + c

# increasing c with small amount
c += h 

# function evaluation after increasing c with a small amount
d2 = a*b + c


print("d1:", d1)
print("d2:", d2)
print("slope:", (d2-d1)/h)
```
